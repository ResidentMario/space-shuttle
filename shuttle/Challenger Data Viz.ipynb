```python
temperatures = [53,57,58,63,66,67,67,67,68,69,70,70,70,72,73,75,75,76,78,79,81]
damage_index = [11,4,4,2,0,0,0,0,0,0,4,0,4,0,0,0,4,0,0,0,0]
# TODO: Also add in dates.
```

```python
plt.output_notebook(hide_banner=True)

p = plt.figure(height=400, width=960)

p.scatter(
    temperatures,
    damage_index
)

plt.show(p)
```

```python
df = pd.DataFrame({'temperature': temperatures, 'damage_index': damage_index})
viz_width = 960
svg_boxed_width = 30
svg_boxed_height = 120
temp_range = [52, 82]
df['x'] = [(viz_width / svg_boxed_width)*(temp - 52) for temp in df['temperature']]
# df['stacked_pos'] = df['temperature'].accumulate()
exist_list = []
y_pos_list = []
# for loc in range(len(df)):
#     row = df.loc[loc]
#     above_row = df.loc[:loc - 1]
#     print(above_row)
for row in df.iterrows():
    temperature = row[1]['temperature']
    y_pos_list.append(exist_list.count(temperature) * svg_boxed_height + 20)
    exist_list.append(temperature)
df['y'] = y_pos_list
```
